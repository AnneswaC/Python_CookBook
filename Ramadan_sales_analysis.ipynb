# EDA

```python
import seaborn as sns
```

```python
plot1=  sns.countplot(x= 'Gender', data= df)
for bars in plot1.containers:
    plot1.bar_label(bars)
```

```python
import matplotlib.pyplot as plt1
```

```python
gender_counts = df['Gender'].value_counts()
# Plot
#plt1.style.use('dark_background')

b= plt1.bar(gender_counts.index, gender_counts.values, width= 0.4,color=['turquoise','red'],alpha=0.9)

#b[1].set_color= 'turquoise'
plt1.xlabel('Gender')
plt1.ylabel('Count')
plt1.title('Count of Male & Female')
for bar, count in zip(b, gender_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), count,
             ha='center', va='bottom', color='black', fontsize=10)
plt1.show()
```

```python
gender_amounts= df.groupby('Gender')['Amount'].sum()
x= gender_amounts.index
y= gender_amounts.values
plt1.bar(x,y, color= ['black','blue'], width=0.3)
plt.xlabel('Gender')
plt1.ylabel('Amount')
plt.title('Amount Spent Based on Gender')
plt.show()
```

From the above chart we can clearly see that, the females had done more shopping than male as they have highest count of order and also highest amount spent for shopping .

# state & zone

```python
a= df.groupby('Zone')['Amount'].mean()
a
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.pie(a.values, labels= a.index, colors= ['blue','turquoise','yellow','green','pink'],startangle=0,autopct='%1.0f%%',shadow=True, explode=[0,0,0,0,0])
plt.show()
```

```python
b= df.groupby('Zone')['Amount'].sum()
b
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.pie(b.values, labels= b.index, colors= ['blue','turquoise','yellow','green','pink'],startangle=0,autopct='%1.0f%%',shadow=True, explode=[0,0,0,0,0])
plt.show()
```

here amount is taken in consideration based on zone. and the pie chart has shown that most order came from central region/zone and and eastern zone is in the last here.. same goes for the amount spent on shopping.

```python
c= df.groupby('State')['Amount'].sum()
c
```

```python
#plt.style.use('dark_background')
top_n = 10
c_top = c.nlargest(top_n)
plt1.figure(figsize=(10,5))
plt1.bar(c_top.index,c_top.values,color=
         ['red','yellow','blue','black','green','turquoise','orange','purple','olive','indigo'] ,width=0.5)
plt1.xlabel('States')
plt1.ylabel('Amount spent')
plt1.xticks(rotation=45) 
#plt1.ylim(0, max(c_top.values) * 1.2)
#plt1.tight_layout()

plt1.show()
```

```python
v= df.groupby('State')['Orders'].sum()
v
```

```python
top_n=10
c1= v.nlargest(top_n)
c1_sorted= c1.sort_values(ascending= False)
plt1.figure(figsize=(10,5))
plt1.bar(c1.index,c1_sorted.values,color=
         ['red','yellow','blue','black','green','turquoise','orange','purple','olive','indigo'] ,width=0.5)
plt1.xlabel('States')
plt1.ylabel('orders_quatity')
plt1.xticks(rotation=45) 
plt1.show()
```

here we can clearly see that Highest order has came from UP followed by MP, karnataka, delhi and all. Also fro previous chart it is visible that the highest amount paid for shopping is from UP and UTTARAKHAND is last in the list.

# Occupation

```python
v1= df.groupby('Occupation')['Amount'].sum()
v1
```

```python
top_n=10
m = v1.nlargest(top_n)
m_sorted= v1.sort_values(ascending= False)
plt1.figure(figsize=(10,5))
plt1.bar(m_sorted.index,m_sorted.values,color=
         ['purple'] ,width=0.5)
plt1.xlabel('Occupation')
plt1.ylabel('Amount')
plt1.xticks(rotation=45) 
plt1.show()
```
